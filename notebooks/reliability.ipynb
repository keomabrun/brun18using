```python
import pandas as pd
```

```python
df = pd.read_csv("../data/hr_neighbors.csv", usecols=range(8))
print "total number of lines =", len(df)

# removing links without communication
df_clean = df[(df.numTxPackets > 0) & (df.numTxFailures >= 0)]
lines_removed = len(df) - len(df_clean)
print "number of lines removed = ", lines_removed, "(", lines_removed*100/len(df), "%)"

# calculating PDR for each link
df_clean.loc[:, 'pdr'] = pd.Series((df_clean.numTxPackets.astype(float) - df_clean.numTxFailures.astype(float)).astype(float) / df_clean.numTxPackets.astype(float), index=df_clean.index)

# indexing by date
df_clean.loc[:, "date"] = pd.to_datetime(df_clean.time, unit='s')
df_clean.set_index(df_clean["date"], inplace=True)

df_clean.head()
```

# PDR and Time

```python
print df_clean["pdr"].describe()

print "Overall PDR =", df_clean.pdr.mean()
print "Total Number of Transmission =", "{:,}".format(df_clean.numTxPackets.sum())
print "Total Number of Failed Transmission =", "{:,}".format(df_clean.numTxFailures.sum())
```

```python
data = []
grouped = df_clean.groupby(["mac"])
for name, group in grouped:
    data.append([name[-8:], len(group.pdr), group.pdr.mean(), group.pdr.std()])
stats = pd.DataFrame(data, columns=["mac", "count", "mean_pdr", "std_dev"])
stats
```
